```python
# convert mp4 to gif
import os
import imageio

path = 'vid' + 'car_gui_rpi_seg'
vid = imageio.get_reader(path + '.mp4',  'ffmpeg')


frames = []
for i, im in enumerate(vid):

    if i % 4 == 0:
        frames.append(im)

print(frames.__len__())

# compress the image

# frames = [im[::2, ::2] for im in frames]


# makes gif loop play



# frames = [im[:, :, ::-1] for im in vid]

imageio.mimsave(path + '.gif', frames, 'GIF', duration=1, loop=0)


```

```python
# jpg 2 gif
import imageio
import os

path = 'vid/bev_pred/'
images = []
for filename in sorted(os.listdir(path), key=lambda x: int(x.split('.')[0])):
    if filename.endswith(".jpg"):
        images.append(imageio.imread(os.path.join(path, filename)))
imageio.mimsave('vid/bev_pred.gif', images, 'GIF', fps=5, loop=0)
```

```python
# mp4 2 img
import cv2
import os

path = 'vid/car_front_1.mp4'

vidcap = cv2.VideoCapture(path)
success,image = vidcap.read()
count = 0
print(success)
while success:
  if count % 5 == 0:
    cv2.imwrite("vid/carla_img/frame%d.jpg" % count, image)     # save frame as JPEG file
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
